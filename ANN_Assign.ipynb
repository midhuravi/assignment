```python
# convert to numpy arrays and split features from labels

x = data.iloc[:,0:20]
y=data.iloc[:,-1]
```

```python
from sklearn.model_selection import GridSearchCV
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import tensorflow as tf
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(x)
```

```python
!pip install scikeras
from scikeras.wrappers import KerasClassifier
```

```python
# Model Creation

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
model=Sequential()
model.add(Dense(10,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=100)
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
y_pred[y_pred>0.5]=1
y_pred[y_pred<=0.5]=0
```

```python
# Evaluating Model Performance with Accuracy Score

from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
# Accuracy obtained is 0.366
```

```python
# Hyperparmeter tuning
def create_model(optimizer='SGD'):
#create a model
  model=Sequential()
  model.add(Dense(12,input_shape=(20,),activation='relu'))
  model.add(Dense(1,activation='sigmoid'))
  model.compile(loss='binary_crossentropy',optimizer=optimizer,metrics=['accuracy'])
  return model
```

```python
# Setting the Random Seed
seed=7
tf.random.set_seed(seed)
```

```python
model=KerasClassifier(model=create_model,batch_size=10,verbose=0)
```

```python
# Defining the Hyperparameter Grid

optimizer = ['SGD',  'Adam', 'Adamax', 'Nadam']
epochs = [10,20,30]
learn_rate = [0.001, 0.01, 0.1, 0.2, 0.3]
param_grid = dict(model__optimizer=optimizer,optimizer__learning_rate=learn_rate,epochs=epochs)
```

```python
#Performing Grid Search for Hyperparameter Tuning

grid=GridSearchCV(estimator=model,param_grid=param_grid,n_jobs=-1,cv=3)
grid_result=grid.fit(x,y)
```

```python
# summarize the results
print("Best: %f using %s" % (grid_result.best_score_,grid_result.best_params_))
meanscore = grid_result.cv_results_['mean_test_score']
stds = grid_result.cv_results_['std_test_score']
params = grid_result.cv_results_['params']
for mean, stdev, param in zip(meanscore, stds, params):
  print("mean= ", mean, "stdev= ",stdev,'with',param)
```
