```python
#Install libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import data file

data=pd.read_csv("/content/train_loan.csv")
data
```

```python
# Checking datatypes

data.info()
```

```python
#Checking missing values

data.isnull().sum()
```

```python
#Filling null values of object datatypes

data['Gender']=data['Gender'].fillna(data['Gender'].mode()[0])
data['Married']=data['Married'].fillna(data['Married'].mode()[0])
data['Dependents']=data['Dependents'].fillna(data['Dependents'].mode()[0])
data['Self_Employed']=data['Self_Employed'].fillna(data['Self_Employed'].mode()[0])
```

```python
# Analysing the nature of distributions of Int and Float datatypes

plt.hist(data['LoanAmount'])
plt.title("Distribution of Loan Amount")
plt.xlabel('Loan Amount')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.hist(data['Loan_Amount_Term'])
plt.title("Distribution of Loan Amount Term")
plt.xlabel('Loan Amount Term')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.hist(data['Credit_History'])
plt.title("Distribution of Credit History")
plt.xlabel('Credit History')
plt.ylabel('Frequency')
plt.show()
```

```python
#Filling null values of int and float

data['LoanAmount']=data['LoanAmount'].fillna(data['LoanAmount'].median())
data['Loan_Amount_Term']=data['Loan_Amount_Term'].fillna(data['Loan_Amount_Term'].median())
data['Credit_History']=data['Credit_History'].fillna(data['Credit_History'].median())
```

```python
data.isnull().sum()
```

```python
#Dropping Irrelavent Columns
data= data.drop(columns=['Loan_ID'])
```

```python
data
```

```python
# Encoding

data.dtypes
```

```python
data.columns
```

```python
cat_cols=['Gender', 'Married', 'Dependents', 'Education',
       'Self_Employed', 'Property_Area', 'Loan_Status']
```

```python
data[cat_cols]
```

```python
data['Dependents']=data['Dependents'].str.replace('+',' ')
data['Dependents']=data['Dependents'].astype('int')
```

```python
x=data.drop('Loan_Status',axis=1)
y=data['Loan_Status']
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y=le.fit_transform(y)
```

```python
x=pd.get_dummies(x,dtype=int)
x
```

```python
# Scaling
from sklearn.preprocessing import MinMaxScaler
min_max=MinMaxScaler()
x.loc[:, 'LoanAmount']=min_max.fit_transform(x[['LoanAmount' ]])

x
```

```python
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
# Logistic Regression

model1=LogisticRegression()
model1.fit(X_train,y_train)
y_pred=model1.predict(X_test)
accuracy=accuracy_score(y_test,y_pred)
```

```python
print("Accuracy score of Logisitic Regression",accuracy*100)
```

```python
score=cross_val_score(model1,x,y,cv=5)
```

```python
score
```

```python
print("Mean score of Logisitic Regression",np.mean(score)*100)
```

```python
#Decision Tree Classifier

model2=DecisionTreeClassifier()
model2.fit(X_train,y_train)
y_pred=model2.predict(X_test)
accuracy=accuracy_score(y_test,y_pred)
print("Accuracy score of Decision Tree Model",accuracy*100)
```

```python
score=cross_val_score(model2,x,y,cv=5)
print("Mean score of Decision Tree Model",np.mean(score)*100)
```

```python
#Random Forest Classifier

model3=RandomForestClassifier()
model3.fit(X_train,y_train)
y_pred=model3.predict(X_test)
accuracy=accuracy_score(y_test,y_pred)
print("Accuracy score of Random Forest Classifier",accuracy*100)
```

```python
#KNeighbor Model

model4=KNeighborsClassifier()
model4.fit(X_train,y_train)
y_pred=model4.predict(X_test)
accuracy=accuracy_score(y_test,y_pred)
print("Accuracy score of KNeighbors Classifier",accuracy*100)
```

```python
score=cross_val_score(model4,x,y,cv=5)
print("Mean score of KNeighborsClassifier",np.mean(score)*100)
```

```python
# Conclusions

# Random Forest Classifier provides the highest accuracy of 76.62.
# Followed by Logistic Regression (75.32) and Decision Tree Classifier (64.94)
# Least accuracy is found in KNeighborsClassifier (57.79)
```
